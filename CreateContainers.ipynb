```python
ibGateways = [['legioIBGW_ASIA1','172.50.0.30','31350'],['legioIBGW_ASIA2','172.50.0.31','31351'],
              ['legioIBGW_ASIA3','172.50.0.32','31352'],['legioIBGW_USA1','172.50.0.33','31353'],
              ['legioIBGW_USA2','172.50.0.34','31354'],['legioIBGW_USA3','172.50.0.35','31355'],
              ['legioIBGW_EUR1','172.50.0.36','31356'],['legioIBGW_EUR2','172.50.0.37','31357'],
              ['legioIBGW_EUR3','172.50.0.38','31358']]
```

```python
for ibgw in ibGateways:
    bashCommand = "sudo docker run -i -t --name "+ibgw[0]+" --net legionet --ip "+ibgw[1]+" -d -p "+ibgw[2]+":5900 legioib"
    process = subprocess.Popen(bashCommand.split(), stdout=subprocess.PIPE)
    output, error = process.communicate()
```

```python
for ibContainer in ibGateways:
    bashCommand = "sudo docker stop "+ibContainer[0]    
    process = subprocess.Popen(bashCommand.split(), stdout=subprocess.PIPE)
    output, error = process.communicate()
    time.sleep(1)
```

```python
for ibContainer in ibGateways:
    bashCommand = "sudo docker start "+ibContainer[0]    
    process = subprocess.Popen(bashCommand.split(), stdout=subprocess.PIPE)
    output, error = process.communicate()
    time.sleep(1)
```
